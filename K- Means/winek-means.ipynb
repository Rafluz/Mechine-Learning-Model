# Membuat Data Training

```python
x_train = dfwine.values
```

```python
x_train
```

# Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

# Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans
```

```python
y_cluster = kmeans.fit_predict(x_train)
y_cluster
```

```python
dfwine['Cluster'] = y_cluster
dfwine
```

```python
kmeans.cluster_centers_
```

# Visualisasi Data Hasil Clustering

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_train)
c_pca = pca.fit_transform(kmeans.cluster_centers_)
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=kmeans.labels_)
plt.scatter(c_pca[:,0],c_pca[:,1],c='red',s=100)
plt.show()
```

# Mencari K (jumlah cluster) terbaik

```python
inertias = []
k_range = range(1,10)
for k in k_range:
    km = KMeans(n_clusters=k,  random_state=42).fit(x_train)
    inertias.append(km.inertia_)
```

```python
inertias
```

```python
plt.xlabel("k")
plt.ylabel("sum of error")
plt.plot(k_range,inertias)
plt.grid()
```
